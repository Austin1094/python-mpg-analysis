```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

## 1.) Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.scatterplot(data=mpg, x='horsepower', y='mpg')

# Adding labels and title
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (mpg)')
plt.title('Relationship between Horsepower and Fuel Economy')

# Displaying the plot
plt.show()
```

## 2.) Create a table summarizing fuel economy by country of origin

```python
# Calculate average fuel economy (mpg) by country of origin
fuel_economy_summary = mpg.groupby('origin')['mpg'].mean().reset_index()

# Display the table
print(fuel_economy_summary)
```

## 3.) Create a visual showing how fuel economy has changed over time

```python
# Plotting the average fuel economy (mpg) over time
sns.lineplot(data=mpg, x='model_year', y='mpg', estimator='mean')

# Adding labels and title
plt.xlabel('Model Year')
plt.ylabel('Average Fuel Economy (mpg)')
plt.title('Change in Fuel Economy Over Time')

# Displaying the plot
plt.show()
```

## 4.) Create two other visuals or tables related to fuel economy.

```python
# Plotting boxplot of fuel economy (mpg) by number of cylinders
sns.boxplot(data=mpg, x='cylinders', y='mpg')

# Adding labels and title
plt.xlabel('Number of Cylinders')
plt.ylabel('Fuel Economy (mpg)')
plt.title('Fuel Economy by Number of Cylinders')

# Displaying the plot
plt.show()
```

```python
# Calculate average fuel economy (mpg) by country of origin
fuel_economy_summary = mpg.groupby('origin')['mpg'].mean().reset_index()

# Plotting barplot of average fuel economy (mpg) by origin
sns.barplot(data=fuel_economy_summary, x='origin', y='mpg')

# Adding labels and title
plt.xlabel('Country of Origin')
plt.ylabel('Average Fuel Economy (mpg)')
plt.title('Average Fuel Economy by Origin')

# Displaying the plot
plt.show()
```
